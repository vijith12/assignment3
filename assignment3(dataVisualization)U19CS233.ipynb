```python
#lineplot seaborn
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
%matplotlib inline
sns.set_theme()
#Loading flights dataset from seaborn
flights = sns.load_dataset("flights")
flights.head()
sns.lineplot(
    data=flights,
   x="year", y="passengers",hue="month"
)
```

```python
#scatter plot
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
%matplotlib inline
sns.set_theme()
#Loading flights dataset from seaborn
flights = sns.load_dataset("flights")
flights.head()
sns.scatterplot(
    data=flights,
   x="year", y="passengers",hue="month"
)
```

```python
#regression plot
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
%matplotlib inline
sns.set_theme()
#Loading flights dataset from seaborn
flights = sns.load_dataset("flights")
flights.head()
sns.regplot(
    data=flights,
   x="year", y="passengers"
)
```
